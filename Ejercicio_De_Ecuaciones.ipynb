```python
import numpy as np
import matplotlib.pyplot as plt

# Ejercicio 1.a
A = np.array([[5, 2],
              [3, -3]])
B = np.array([1, 5])

sol = np.linalg.solve(A, B)
print("Ejercicio 1.a solución (x, y):", sol)

x_vals = np.linspace(-10, 10, 200)
y1 = (1 - 5*x_vals) / 2
y2 = (3*x_vals - 5) / 3

plt.plot(x_vals, y1, label="5x + 2y = 1")
plt.plot(x_vals, y2, label="3x - 3y = 5")
plt.scatter(sol[0], sol[1], color="black", label=f"Solución {sol}")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Ejercicio 1.a")
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Ejercicio 4.b
A = np.array([[2, 3],
              [-6, 12]])
B = np.array([2, 1])

rankA = np.linalg.matrix_rank(A)
rankAb = np.linalg.matrix_rank(np.c_[A, B])

if rankA == rankAb == A.shape[1]:
    sol = np.linalg.solve(A, B)
    print("Ejercicio 4.b solución (x, y):", sol)
else:
    print("Ejercicio 4.b: no tiene solución única.")

x_vals = np.linspace(-10, 10, 200)
y1 = (2 - 2*x_vals) / 3
y2 = (1 + 6*x_vals) / 12

plt.plot(x_vals, y1, label="2x + 3y = 2")
plt.plot(x_vals, y2, label="-6x + 12y = 1")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Ejercicio 4.b")
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Ejercicio 8.a
A = np.array([[3, -2],
              [2,  1]])
B = np.array([-4, 2])

sol = np.linalg.solve(A, B)
print("Ejercicio 8.a solución (x, y):", sol)

x_vals = np.linspace(-10, 10, 200)
y1 = (3*x_vals + 4) / 2
y2 = 2 - 2*x_vals

plt.plot(x_vals, y1, label="3x - 2y = -4")
plt.plot(x_vals, y2, label="2x + y = 2")
plt.scatter(sol[0], sol[1], color="black", label=f"Solución {sol}")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Ejercicio 8.a")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Ejercicio 8.b
A = np.array([[1, -4],
              [3, -12]])
B = np.array([5, 15])

rankA = np.linalg.matrix_rank(A)
rankAb = np.linalg.matrix_rank(np.c_[A, B])

if rankA == rankAb == A.shape[1]:
    sol = np.linalg.solve(A, B)
    print("Ejercicio 8.b solución (x, y):", sol)
elif rankA == rankAb:
    print("Ejercicio 8.b: infinitas soluciones.")
else:
    print("Ejercicio 8.b: sistema incompatible (sin solución).")

x_vals = np.linspace(-10, 10, 200)
y1 = (x_vals - 5) / 4
y2 = (3*x_vals - 15) / 12

plt.plot(x_vals, y1, label="x - 4y = 5")
plt.plot(x_vals, y2, label="3x - 12y = 15")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Ejercicio 8.b")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_vals = np.linspace(-5, 5, 100)
y_vals = 10 + 6*x_vals

puntos = [(0,10), (1.5,19), (-1,-4), (0,0.4), (-0.5,7)]
soluciones = []
for punto in puntos:
    x, y = punto
    if abs(-3*x + 0.5*y - 5) < 1e-5:
        soluciones.append(punto)

plt.plot(x_vals, y_vals, label='-3x + 0.5y = 5')
for punto in puntos:
    plt.scatter(punto[0], punto[1], color='blue')
for sol in soluciones:
    plt.scatter(sol[0], sol[1], color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ejercicio nº 20')
plt.legend()
plt.grid(True)
plt.show()
```
