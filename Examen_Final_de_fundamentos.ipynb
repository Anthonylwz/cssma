Parte 3: Algoritmia aplicada

Escribe una función en Python que reciba una lista de números y devuelva:
o El número mayor
o El número menor
o El promedio

```python
def analyze_list(numbers):
  """
  Analiza una lista de números y devuelve: el mayor, el menor y el promedio.

  Args:
    numbers: Una lista de números.

  Returns:
    Una tupla que contiene el número mayor, el número menor y el promedio.
  """
  if not numbers:
    return None, None, None

  largest = max(numbers)
  smallest = min(numbers)
  average = sum(numbers) / len(numbers)

  return largest, smallest, average

my_list = [10, 5, 20, 15, 25]
largest_num, smallest_num, avg_num = analyze_list(my_list)

print(f"Número mayor: {largest_num}")
print(f"Número menor: {smallest_num}")
print(f"Promedio: {avg_num}")
```

Usando NumPy, implementa una función de activación Sigmoide definida como:
f(x)=11+e−xf(x) = \frac{1}{1 + e^{-x}}f(x)=1+e−x1
Aplica la función a un arreglo x = np.array([-2, -1, 0, 1, 2])

```python
import numpy as np

def sigmoid(x):
  """
  Implementa la función de activación Sigmoide.

  Args:
    x: Un arreglo NumPy.

  Returns:
    El arreglo con la función Sigmoide aplicada elemento a elemento.
  """
  return 1 / (1 + np.exp(-x))

x = np.array([-2, -1, 0, 1, 2])

sigmoid_output = sigmoid(x)

print("Arreglo de entrada:", x)
print("Salida de la función Sigmoide:", sigmoid_output)
```

Crea una función que clasifique a los estudiantes del DataFrame en "Aprobado"
si su Promedio ≥ 14, y "Desaprobado" en caso contrario. Añade una nueva
columna llamada Estado.

```python
def classify_student(promedio):
  """
  Clasifica a un estudiante como "Aprobado" o "Desaprobado" basado en su promedio.

  Args:
    promedio: El promedio del estudiante.

  Returns:
    "Aprobado" si el promedio es mayor o igual a 14, de lo contrario "Desaprobado".
  """
  if promedio >= 14:
    return "Aprobado"
  else:
    return "Desaprobado"

df_students['Estado'] = df_students['Promedio'].apply(classify_student)

display(df_students)
```

crecimiento entre Ventas_2024 y Ventas_2025.

```python

datos = {
    'Producto': ['A', 'B', 'C', 'D', 'E'],
    'Ventas_2024': [120, 340, 560, 230, 150],
    'Ventas_2025': [200, 400, 600, 300, 180]
}

df = pd.DataFrame(datos)

df['Crecimiento'] = ((df['Ventas_2025'] - df['Ventas_2024']) / df['Ventas_2024']) * 100

print(df)
```

Ecuación de primer grado

ax+b=0ax + b = 0ax+b=0
por ejemplo: 2x + 4 = 0.
o La función debe retornar el valor de x.

```python
import numpy as np

# Ejercicio 1
# Ecuación: 2x + 4 = 0
# Se representa como sistema: [2] * x = -4
A = np.array([[2]])
B = np.array([-4])

sol = np.linalg.solve(A, B)
print("Ejercicio 1 solución (x):", sol)
```

```python
import numpy as np

# Ejercicio 2
# Ecuación: x^2 + 5x + 6 = 0
a, b, c = 1, 5, 6

# Fórmula general
discriminante = b**2 - 4*a*c
x1 = (-b + np.sqrt(discriminante)) / (2*a)
x2 = (-b - np.sqrt(discriminante)) / (2*a)

print("Ejercicio 2 soluciones (x1, x2):", (x1, x2))
```
