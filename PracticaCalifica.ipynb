Concatenar dos vectores: vertical y horizontalmente

```python
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])

concat_horizontal = np.concatenate((vector1, vector2))
print("\nconcatenacion horizontal:\n", concat_horizontal)

vector1_col = vector1.reshape(-1, 1)
vector2_col = vector2.reshape(-1, 1)
concat_vertical = np.concatenate((vector1_col, vector2_col), axis=0)
print("concatenacion vertical:\n", concat_vertical)
```

Ejercicio 6.5

Agrupar por ciudad y calcular el promedio de edad

```python
import pandas as pd

# Recreate the DataFrame with all necessary columns
data = {'Nombre': ['Juan', 'Mara', 'Pedro', 'Ana', 'Luis'],
        'Edad': [28, 22, 35, 29, 20],
        'Ingresos': [2000, 2500, 2300, 1800, 2700],
        'Ciudad': ['Madrid', 'Barcelona', 'Madrid', 'Valencia', 'Sevilla']}
df = pd.DataFrame(data)

promedio_por_ciudad = df.groupby('Ciudad')['Edad'].mean()
print("\n Promedio de edad por ciudad:\n", promedio_por_ciudad)
```
